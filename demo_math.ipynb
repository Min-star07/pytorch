```python
import torch
```

```python
a = torch.rand(2, 3)
print(a)
```

```python
print(torch.cos(a))
print(torch.sin(a))
```

```python
# torch.manual_seed(1)
a = torch.rand(2, 3)
print(torch.normal(mean=a, std=0.1))
```

```python
a = torch.rand(1, 1)
print(a)
print(torch.norm(a, p=2))
print(torch.norm(a, p=1))
```
